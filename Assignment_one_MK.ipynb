Can you sort a dictionary? Why or why not?<br><br>


No, you cannot directly sort a dictionary in Python because dictionaries are inherently unordered data structures. The elements in a dictionary are stored as key-value pairs, and their order is not preserved.

However, if you need to retrieve the items from a dictionary in a specific order, you can sort the keys and then access the corresponding values.

## Sets 

What is unique about a set?<br><br>


Sets in Python are unordered collections of unique elements. This means that a set cannot contain duplicate values. If you try to add an element to a set that already exists in the set, it will not be added again.

Use a set to find the unique values of the list below:

````python
list5 = [1,2,2,33,4,4,11,22,3,3,2]
unique_values = set(list5)
print(unique_values) 
````
